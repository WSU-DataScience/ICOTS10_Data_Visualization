# Altair: Statistical Visualization for Python

* Declarative API for statistical visualization, 
* built on top of [Vega-Lite](http://vega.github.io/vega-lite/).
* **Link:** [Altair](http://github.com/altair-viz/altair/)

## Importing `altair`

```python
import altair as alt
from vega_datasets import data
cars = data.cars()
```

```python
# Uncomment/run this line to enable Altair in the classic notebook
#   (this is not necessary in JupyterLab)
#alt.renderers.enable('notebook')
```

## Quick Altair example

```python
chart = alt.Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)

chart
```

## Why use `Altair`

* Power of `vega-lite`
    * JavaScript
    * Create complicated, interactive plots
* Easier syntax and prototyping
* Easy export to html
    * No server needed!
